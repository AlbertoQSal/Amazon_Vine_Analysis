```python
# Import dependencies
import pandas as pd
```

```python
data = pd.read_csv('vine_table.csv')
df = pd.DataFrame(data)
df
```

```python
df.dtypes
```

```python
# Filter the data and create a new DataFrame  where the total_votes count is equal to or greater than 20
totalVotes_df = df.loc[df['total_votes']>=20]
totalVotes_df
```

```python
# Filter the totalVotes_df to retrieve all the rows where the number of helpful_votes divided 
# by total_votes is equal to or greater than 50%.
votesPercentage_df = totalVotes_df.loc[(df['helpful_votes']/df['total_votes'])>=0.5]
votesPercentage_df
```

```python
# Filter the votesPercentage_df and retrieve all the rows where a review 
# was written as part of the Vine program (paid), vine == 'Y'.
reviewY_df = votesPercentage_df.loc[votesPercentage_df['vine'] == 'Y']
reviewY_df
```

```python
# Filter the votesPercentage_df and retrieve all the rows where a review 
# was written as part of the Vine program (unpaid), vine == 'N'.
reviewN_df = votesPercentage_df.loc[votesPercentage_df['vine'] == 'N']
reviewN_df
```

```python
# Determine: total number of reviews, the number of 5-star reviews, and the 
# percentage of 5-star reviews for the two types of review (paid vs unpaid).

total_5Stars_reviews = total_5Stars_reviews = df[df['star_rating']==5]['star_rating'].count()

results = ({
    'Total_Number_of_Reviews' : len(df),
    'Number_of_5Stars' : df[df['star_rating'] == 5]['star_rating'].count(),
    'Percentage_5StarsReviews_Paid' : (df[(df['star_rating']==5) & (df['vine']=='Y')]['vine'].count()/total_5Stars_reviews)*100,
    'Percentage_5StarsReviews_Unpaid' : ((df[(df['star_rating']==5) & (df['vine']=='N')]['vine'].count())/total_5Stars_reviews)*100
})

results

```
